```python
from time import time
from chespex.molecules import MoleculeGenerator
import matplotlib.pyplot as plt
```

```python
max_number_of_beads_per_molecule = 4
bead_types = [
    'Q5+', 'Q4+', 'Q3+', 'Q2+', 'Q1+',
    'Q5-', 'Q4-', 'Q3-', 'Q2-', 'Q1-',
    'P6', 'P5', 'P4', 'P3', 'P2', 'P1',
    'N6', 'N5', 'N4', 'N3', 'N2', 'N1',
    'C6', 'C5', 'C4', 'C3', 'C2', 'C1',
    'X4', 'X3', 'X2', 'X1'
]
molecule_generator = MoleculeGenerator(max_beads=max_number_of_beads_per_molecule, bead_types=bead_types)
```

We iterate over all molecules and thereby count them:

```python
def count_molecules():
    molecule_count = 0
    for _ in molecule_generator.generate():
        molecule_count += 1
    return molecule_count

start_time = time()
molecule_count = count_molecules()
end_time = time()

print(f'Generated {molecule_count:,} molecules in {end_time - start_time:.2f} seconds')
```

Now, we plot a few molecules to see what they look like.

```python
fig, axs = plt.subplots(5, 5, figsize=(12, 12))
plot_count = 0
for i, molecule in enumerate(molecule_generator.generate()):
    if i % 6543 == 0 and i > 90: # Arbitrary selection criteria
        ax = axs[plot_count // 5, plot_count % 5]
        molecule.visualize(ax=ax)
        ax.set_title(f'Molecule {i}')
        ax.margins(0.1)
        plot_count += 1
    if plot_count == 25:
        break
plt.show()
```
